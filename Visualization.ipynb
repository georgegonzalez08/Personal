# creating bar charts

```python
bb = data.DataReader(name="BB", data_source="stooq", start="2007-01-01", end = "2020-12-31")
bb.head(3)
```

```python
bb["Close"]
```

```python
def rank_performance(stock_price):
    if stock_price <=10:
        return "Poor"
    elif stock_price <=50:
        return "Satisfactor"
    else:
        return"Stellar"
```

```python
#applying function to all column values
bb["Close"].apply(rank_performance)
```

```python
#most of the time, it was poor when it closed
#barh makes it horizontal
#bb["Close"].apply(rank_performance).value_counts().plot(kind="bar")
bb["Close"].apply(rank_performance).value_counts().plot(kind="barh")
```

# creating pie charts to represent proportions

```python
bb = data.DataReader(name="BB", data_source="stooq", start="2007-01-01", end = "2020-12-31")
bb.head(3)
```

```python
bb["Close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 29.451:
        return "above average"
    else:
        return "below average"
```

```python
plt.style.use("dark_background")
bb["Close"].apply(rank_performance).value_counts().plot(kind="pie",legend=True)
```
