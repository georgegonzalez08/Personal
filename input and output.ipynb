# export excel file

```python
url = "https://data.cityofnewyork.us/resource/25th-nujf.csv?$query=SELECT%20%60brth_yr%60%2C%20%60gndr%60%2C%20%60ethcty%60%2C%20%60nm%60%2C%20%60cnt%60%2C%20%60rnk%60%0AORDER%20BY%0A%20%20%60brth_yr%60%20DESC%20NULL%20LAST%2C%0A%20%20%60gndr%60%20ASC%20NULL%20LAST%2C%0A%20%20%60ethcty%60%20ASC%20NULL%20LAST%2C%0A%20%20%60rnk%60%20ASC%20NULL%20LAST"
baby_names = pd.read_csv(url)
baby_names.head()
```

```python
girls = baby_names[baby_names["gndr"] == "FEMALE"]
boys = baby_names[baby_names["gndr"] == "MALE"]
```

```python
girls
```

```python
boys
```

```python
#first create excel writer object. creates the container/base to load the files into
excel_file = pd.ExcelWriter("Baby_Names.xlsx")
```

```python
#if dont specify columns, it will include all columns
girls.to_excel(excel_file,sheet_name="Girls", index=False,encoding="utf-8")
```

```python
boys.to_excel(excel_file,sheet_name="Boys", index=False,columns=["brth_yr","gndr","ethcty"],encoding="utf-8)
```

```python
#this is the point, where it actually saves the excel
excel_file.close()
```
