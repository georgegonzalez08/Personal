# drop rows with null values

```python
nba = pd.read_csv(filepath_or_buffer='Pandas/nba.csv')
nba.head(25)
```

```python
#will remove the entire row if there is nan at all in any column
nba.dropna()
```

```python
#only removes entire row if all columns in the row have a nan value
nba.dropna(how='all')
```

```python
#remove a row if we have an nan value in a certain column. remove any rows where college value is nan. 
nba.dropna(subset=['College'])
```

```python
#remove a row if null in two or more columns. 
#has to be null in both, if null, then it will be removed
#check for null values in either column. 
nba.dropna(subset=['College','Salary'])
```

# fill in missing dataframe values with the fillna method 

```python
nba = pd.read_csv(filepath_or_buffer='Pandas/nba.csv')
nba.head(25)
```

```python
#replace all nans with 0
nba.fillna(0)
```

```python
#targeting a column and filling in values. doing this does not make the change permanent to the df. 
nba["College"].fillna("Unknown")
#nba.fillna(["College"])
```

```python
#targeting a column and filling in values. doing this with inplace,makes it permanent to the df. 
nba["College"].fillna("Unknown",inplace=True)
nba.head()
```

```python
#can use this second option as the better option. use new series to overwrite the original series.
nba["College"] = nba["College"].fillna("Unknown")
nba.tail()
```

```python
nba["Salary"] = nba["Salary"].fillna(value = 0)
```

```python
nba.tail()
```

# rank values with the rank method

```python
#removing nan rows
nba = pd.read_csv("pandas/nba.csv").dropna(how="all")
nba.head()
```

```python
#first making nan a 0 in salary column and overwriting. and int values
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
#tells you the rank or the position of the salary based on all the other salaries sorted.
#by default, its set to true, lowest value is rank 1 and highest salary is max rank
nba["Salary"].rank()
```

```python
#number 1 rank is the salary that is the highest
#adding a new column to df with the ranking.
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
```

```python
nba
```

```python
#now sorted by highest salary rank and a part of data frame
nba.sort_values(["Salary"],ascending=False)
```
