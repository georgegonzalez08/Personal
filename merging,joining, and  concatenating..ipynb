# outer joins

```python
week1 = pd.read_csv("Pandas/Restaurant - Week 1 Sales.csv")
week2 =  pd.read_csv("Pandas/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Pandas/Restaurant - Customers.csv")
food = pd.read_csv("Pandas/Restaurant - Foods.csv")
```

```python
#wherever there is a nan, value doesn't exist in either dataframe. 
#displays the matches in both and also ones that don't exist in both.
#customer id column is named the same in both tables
#full list of customers
#indicator creates a new column and where its pulled from
merged = week1.merge(week2,how="outer",on="Customer ID",suffixes=[" - Week 1"," - Week 2"],
           indicator=True)
merged.head()
```

```python
merged["_merge"].value_counts()
```

```python
merged["Customer ID"].nunique()
```

```python
#only pull where they exist separately in each table but not in both
mask1 = merged["_merge"].isin(["left_only","right_only"])
```

```python
merged[mask1]
```

# the left on and right on parameters


```python
week1 = pd.read_csv("Pandas/Restaurant - Week 1 Sales.csv")
week2 =  pd.read_csv("Pandas/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Pandas/Restaurant - Customers.csv")
food = pd.read_csv("Pandas/Restaurant - Foods.csv")
```

```python
# id in customers is the same as customer id customers
customers.head(3)
```

```python
#customer id in week2 is the same as id customers
week2.head(2)
```

```python
#joining on id when columns have two different values in each table
#sorts on customer id
#drop is used to remove the extra ID column.
week2 = week2.merge(customers,how="left",left_on = "Customer ID",right_on = "ID",sort=True).drop("ID",axis="columns")
#wee.merge(food,how="inner",on="food ID")
```

# merging by indexes with the left_index and right_index parameters

```python
week1 = pd.read_csv("Pandas/Restaurant - Week 1 Sales.csv")
week2 =  pd.read_csv("Pandas/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Pandas/Restaurant - Customers.csv",index_col="ID")
food = pd.read_csv("Pandas/Restaurant - Foods.csv",index_col="Food ID")
```

```python
customers
```

```python
week1.head(2)
```

```python
#matches customer id column from week1 df to index of customers.
#dont need to state the column name because it is the index
#duplicate columns dont come up here like above, which we removed the extra id column
sales = week1.merge(customers,how="left",left_on="Customer ID",right_index=True)
```

```python
sales
```

```python
food
```

```python
#matches sales food id column to the index of the food df
sales = sales.merge(food,how="left",left_on="Food ID",right_index=True)
```

```python
sales.head(3)
```

```python
week1.head(3)
```

```python
week2.head(3)
```

```python
#join tables by their indexes. common index between them
#join on left week1 index adn rigth week2 index
week1.merge(week2,how="left",left_index=True,right_index=True,suffixes=[" -week1"," -week2"])
```
