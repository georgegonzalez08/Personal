# filtering with the where method

```python
bond = pd.read_csv("pandas/jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#extract rows where actor is sean connery
mask = bond["Actor"] == "Sean Connery"
```

```python
#only contain rows where sean connery is actor
bond[mask]
```

```python
#returns entire df. but only rows with sean connery has values.
bond.where(mask)
```

```python
bond.where(bond["Box Office"] > 800)
```

```python
mask2 = bond["Box Office"] > 800
```

```python
bond.where(mask & mask2)
```

# the apply method to a column

```python
bond = pd.read_csv("pandas/jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def convert_to_string_and_millions(number):
    return str(number) + " millions!"
bond["Box Office"] = bond["Box Office"].apply(convert_to_string_and_millions)
bond["Box Office"]
```

```python
 bond["Budget"] = bond["Budget"].apply(convert_to_string_and_millions)
```

```python
#instead of above , can do it this way too.
#don't have to write it out 3 times like before. through a for loop
columns =["Box Office","Budget","Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_millions)
bond.head(2)
```

# the .apply() method with row values

```python
bond = pd.read_csv("pandas/jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
#create a new column based on other row values
#goes through each row, based on the column values, and then creates a new column based on them
#doesn't write to dataframe.
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor =="Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget >40:
        return "Enjoyable"
    else:
        return "I have no idea"
        
bond.apply(good_movie,axis="columns")
```
